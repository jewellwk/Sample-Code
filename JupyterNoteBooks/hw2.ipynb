### Answer to Part D

```python
import math
data = movie_ratings.values.tolist()
cosvalues = np.ones((10,10))
n = len(data)
k = 0
while(k < n-1):
    for i in range(n-1-k):
        dotsum = 0
        len_one = 0
        len_two = 0
        for j in range(7):
            dotsum += data[k][j]*data[i+k+1][j]
            len_one += data[k][j]**2
            len_two += data[i+k+1][j]**2
        calc = math.acos(dotsum/(math.sqrt(len_one)*math.sqrt(len_two)))
        cosvalues[i+k+1][k] = '{0:.2f}'.format(math.degrees(calc))
    k+=1
coschart = pd.DataFrame(cosvalues, index=['June', 'Jeff', 'Jerome', 'Judy', 'Jake', 
                                    'Jill', 'Julie', 'Jeremiah', 'Janice', 'Josiah']
                 , columns=['June', 'Jeff', 'Jerome', 'Judy', 'Jake', 
                                    'Jill', 'Julie', 'Jeremiah', 'Janice', 'Josiah'])
modchart = coschart.replace(1.00, math.inf)
person1 = modchart.min(axis=0).idxmin()
person2 = modchart.min(axis=1).idxmin()
print(person1,'and',person2)
```
